```python
import pandas as pd
import numpy as np
data=pd.read_csv('/cars.csv')
print(data)
```

```python
data['miles'].fillna("null",inplace=True)
print(data)
```

```python
data = data.dropna(axis = 1
                   , how ='any')
print(data)
```

```python
import seaborn as s 
import matplotlib.pyplot as plt
s.histplot(x='cars',data=data)
plt.show()
```

```python
s.boxplot(x='age',y='miles',data=data)
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot( x='age', y='miles', data=data,color='green')

plt.show()
```
